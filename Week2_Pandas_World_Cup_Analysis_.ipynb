# Análisis de filas Nulas
La función isnull() nos permite validar si existe información nula o nan en una columna en particular.  Pero en este caso y después de validar el dataset con la función tail(); se estableció que un registro se puede determinar inválido si todas sus columnas son nulas. Para hacer esta búsqueda agregamos el método all() al final de la función isnull() con lo cuál se creará una nueva columna boleena que nos indicará cuales son las filas que se encuentran nulas.

````python
# elimina filas nulas en el conjunto de datos
df_partidos=df_partidos.dropna(how='all')
````

La funcioón dropna nos permite eliminar los registros nulos de un conjunto de datos, el parámetro **"how"** permite definir la regla de eliminación de las filas.  ***All*** eliminará la fila si todas las columnas son nulas, y ***any*** lo realizará si alguna columna en la fila es nula. Ahora que hemos eliminado los registros nulos con la función df_partidos.dropna(how='all'). 

````python
check_nan =df_partidos.isnull().all()
````

# Renombre de columnas
Una de las tareas más comunes en el análisis de datos es realizar un renombrado de las columnas en el dataset. Para ello podemos utilizar la función replace y la propiedad columns de nuestro dataframe.  con el comando for recorreremos todas las columnas. Aquí estamos reemplazando los espacios en blanco por "_"

````python
df_partidos.columns = [c.replace(' ', '_') for c in df_partidos.columns]
````

# **iloc y loc**
Las funciones iloc y loc nos permiten filtrar la información de un conjunto de datos, por ejemplo en el caso anterior, con la función max() recibíamos un valor escalar correspondiente al máximo año de copa mundo en el conjunto de datos y esta entrada la utilizamos como fuente para filtrar los partidos que se jugaron en dicha copa. ***Loc*** nos permite filtrar por un campo en particular, mientras que ***iloc***, nos permite filtrar por el indexador del conjunto de datos.

# **Filtros por campos de texto**
En los conjuntos de datos que tienen cadenas de textos debemos tener algo de cuidado, porque primero debemos covertir el tipo de datos de la columna a texto y luego si proceder a usar las funciones de comparación de texto como contains para asegurarnos que el valor que estamos buscando se encuentre. 

# **Filtros compuestos**
Existe también la posibilidad de hacer filtros compuestos lo cual permite avanzar en los análisis de nuestros conjuntos de datos, el siguiente ejemplo nos permite conocer que partidos se jugaron en montevideo en el mundial de 1930 y donde uno de los equipos fue Uruguay.



````python
#Ejemplo filtros compuestos
df_partidos.loc[(df_partidos["City"].str.contains("Montevideo")) 
                &  (df_partidos["Year"]==(df_copas["Year"].min())) 
                &  ((df_partidos['Home_Team_Name'].str.contains("Uruguay")) | (df_partidos['Away_Team_Name'].str.contains("Uruguay")))]
````
